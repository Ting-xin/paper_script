```python
import pandas as pd
import xarray as xr
import h3
import geojson
```

```python
# Replace with your CSV file name
file_name = 'nanjing_poi.csv'

# Load CSV file into Pandas DataFrame
df = pd.read_csv(file_name)
```

```python
df['hex'] = pd.DataFrame(df['hex7'].apply(
    lambda x: geojson.Polygon([h3.h3_to_geo_boundary(x)])), index=df.index)
df['hex'] = df['hex'].astype(str)

ds = xr.Dataset.from_dataframe(df)
ds.to_netcdf('h3_data.nc')
```
