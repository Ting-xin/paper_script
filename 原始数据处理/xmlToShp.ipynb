```python
# ogr2ogr --config SHAPE_ENCODING "GBK" -f "ESRI Shapefile" 5.shp test.xml -a_srs WGS84
import os
import glob
import re
import subprocess

src_dir = r'E:\paper\data\原始数据\历年行政区划数据\county_xml'

# 批量转换 shp
files = glob.glob(src_dir + '\\*.xml', recursive=True)
for file in files:
  print('start: ', file)
  # print(re.findall(r'\d+', file)[0])
  # print(src_dir.replace('xml', 'shp') + '\\' + re.findall(r'\d+', file)[0])
  cmd = ' '.join(['ogr2ogr', '--config', 'SHAPE_ENCODING', '"GBK"', '-f "ESRI Shapefile"', src_dir.replace('xml', 'shp') + '\\' + re.findall(r'\d+', file)[0] + '.shp', file, '-a_srs', 'WGS84'])
  # print(cmd)
  p = subprocess.Popen(cmd, shell=True)
  p.wait()
  print('success')




```
