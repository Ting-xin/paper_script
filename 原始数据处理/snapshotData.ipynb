```python
import glob
import json
import re
import hashlib
import json

def gen_md5_id(item):
    md5_machine = hashlib.md5()
    md5_machine.update(item.encode('utf-8'))
    return md5_machine.hexdigest()

src_dir = r'E:\paper\data\原始数据\历年行政区划数据\1949-2014伯克利数据\province_geojson'

files = glob.glob(src_dir + '\\*', recursive=True)

result = []   # 每个 item 增加一个生命周期，每碰见一个新的item，就加入result，并且将该生命期作为启示，如果该年没有这个item，则表示生命期结束
pre_year = {}
now_year = {}
len_all = 0
for file in files:
  year = re.compile('\d+\.json').findall(file)[0].replace('.json', '')
  year = int(year)
  with open(file, encoding='utf-8') as f:
    data = json.load(f).get('features')
    len_all += len(data)
    now_year = {}
    for i in data:
      properties = i.get('properties')
      pcode = properties['PCODE']
      if 'PROV_CH' in properties:
        name = properties['PROV_CH']
      else:
        name = properties['PROV_CN']
      geometry = i.get('geometry')
      id = gen_md5_id(str(pcode) + str(name) + str(geometry))

      temp = {}
      temp['name'] = name
      temp['code'] = pcode
      temp['geometry'] = geometry
      now_year[id] = temp
    
    temp = {}

    for i in pre_year:
      if i not in now_year:
        pre_year[i]['eYear'] = year
        result.append(pre_year[i])
      else:
        temp[i] = pre_year[i]
    pre_year = temp
    
    for i in now_year:
      if i not in pre_year:
        now_year[i]['sYear'] = year
        pre_year[i] = now_year[i]


for i in pre_year.values():
  if 'eYear' not in i :
    i['eYear'] = 2015
  result.append(i)

with open(r'E:\paper\data\原始数据\历年行政区划数据\1949-2014伯克利数据\province.json', 'w', encoding='utf-8') as f:
  json.dump(result, f, ensure_ascii=False,indent = 4)

```
