```python
import geopandas as gpd
import h3pandas
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']  # 
plt.rcParams['axes.unicode_minus'] = False  # 

gdf = gpd.read_file('.json')
ax = gdf.plot(color='#eed8aa', figsize=(10,10))
gdf_h3 = gdf.h3.polyfill_resample(3)
gdf_h3.plot(color='grey', ax=ax, alpha=0.3)
ax.set_xticks([])
ax.set_yticks([])
plt.axis('off')

```
